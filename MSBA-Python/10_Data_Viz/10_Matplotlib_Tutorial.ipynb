### Basic Matplotlib Plots

```python
## common convention to import as plt so we don't have to write out pyplot every time
from matplotlib import pyplot as plt

## easier than calling each one every time e.g. matplotlib.pyplot.plot() 
```

```python
## Importing other libraries we'll use
import numpy as np
import pandas as pd
```

```python
## Values for x-axis
Xvalues = [2, 4, 6, 8]

## Values for y-axis
Yvalues = [1, 2, 3, 4]
```

```python
plt.plot(Xvalues)
```

#### Per Matplotlib Documentation:
If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. 

#### Line Plots
Syntax for line plots is plt.plot(x,y), where x and y are arrays of the same length that specify the (x, y) pairs that form the line

```python
## first our x-value, then our y-values (syntax = plt.plt(x,y))
plt.plot(Xvalues, Yvalues)
# plt.show()  ## use to remove chart data
```

```python
# Note: in other coding environments you may need to run the lines below to get the chart to show
# plt.show()
```

```python
## Adding in some plot info

plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.title('Testing out Basic Plotting')
plt.plot(Xvalues, Yvalues, label = 'Our Line')
plt.legend()
```

```python
### More realistic data & adding in some formatting

age = [18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42, 44, 46, 48, 50, 52, 54]

salary = [17100, 20000, 30500, 41247, 48876, 57287, 65998, 70000, 75370, 84666,
            78254, 87038, 91991, 94796, 97962, 99240, 102736, 112285, 120000]

## can play around with all of these formatting settings
plt.plot(age, salary, color='green', linestyle='--', marker = 'o', label='Salary')
plt.legend()
```

```python
## More Styles (styles you can use that are built into Matplotlib)
plt.style.available
```

```python
s_masters = []

for item in salary:
    item = item*1.10
    s_masters.append(item)
```

```python
# plotting the points 
plt.plot(age, salary, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=5, label = 'Base Salary')

plt.plot(age, s_masters, color='yellow', linestyle='dotted', linewidth = 3,
         marker='v', markerfacecolor='red', markersize=10, label = 'Salary with Masters')

plt.legend()

plt.style.use('seaborn-dark')

### naming the x & y axes
plt.xlabel('Age (Years)')
plt.ylabel('Salary ($)')

  
### setting x and y axis range
plt.ylim(1,150000)
plt.xlim(10,60)
  
```

### Barplots

Basics of the barplot:

```python
# x-values of left sides of bars 
xvals = [1, 2, 3, 4, 5]
  
# heights of bars
height = [10, 24, 36, 40, 5]
  
# labels for bars
ticks = ['one', 'two', 'three', 'four', 'five']
  
# plotting a bar chart
plt.bar(xvals, height, tick_label = ticks,
        width = 0.5, color = 'red') ## test adding in different colors
  
# naming the x-axis
#plt.xlabel('Bars')
# naming the y-axis
plt.ylabel('Height')
# plot title
plt.title('Testing Out Bar Charts')
```

```python
## Revisiting previous data but plotting bar charts vs line charts
plt.bar(age, salary, label = 'Base Salary', color = 'black')
plt.bar(age, s_masters, label='Salary with Masters', color = 'green')

## we can't actually see the base salary values because they are covered
```

```python
## One solution:

x_indexes = np.arange(len(age)) ## giving age values an index and making them an array
width = .40 

plt.bar(x_indexes - width, salary, label = 'Base Salary', width = width, color = 'black')
plt.bar(x_indexes, s_masters, label='Salary with Masters', width = width, color = 'green')

plt.legend()

## But labels are now wrong, so we reformat them to match the Index but show Agee
plt.xticks(ticks = x_indexes, labels = age);
```

### Histograms

```python
testScores = [65, 69, 72, 74, 78, 81, 83, 86, 90, 92, 99]
```

```python
fig = plt.figure(figsize=(10, 8))

fig_width, fig_height = plt.gcf().get_size_inches()
print(fig_width)
print(fig_height)

## Can also pass in our own values for the bins
bins = [60, 65, 70, 75, 80, 85, 90, 95, 100]

plt.hist(testScores, bins=bins, edgecolor='black', color='white')

import statistics
median_score = statistics.median(testScores)

plt.axvline(median_score, color='red', label='Median Score', linewidth=2)
plt.legend()

plt.title('Test Scores')
plt.xlabel('Scores')
plt.ylabel('Total Respondents')

plt.tight_layout()  ## add some padding to make it look better

## Changing the ticks:
plt.yticks([0,1,2])
plt.yticks(np.arange(0, 3, 1)) 

## Can also add different. labels:
plt.yticks(np.arange(3), ['Bad', 'Better', 'Best'], rotation=20) 

## Or make them go away entirely:
plt.xticks([])  # Disable xticks.
```

### Subplots

Groups of axes that can exist in a single matplotlib figure.

```python
plt.gcf()  ## get current figure i.e. gives the reference of the current figure (since there is none it creates them)
plt.gca()  ## get current axes i.e. gives the reference of the current axes
```

```python
## Set the style
plt.style.use('seaborn')

age = [18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42, 44, 46, 48, 50, 52, 54]

salary = [17100, 20000, 30500, 41247, 48876, 57287, 65998, 70000, 75370, 84666,
            78254, 87038, 91991, 94796, 97962, 99240, 102736, 112285, 120000]

s_masters = []

for item in salary:
    item = item*1.10
    s_masters.append(item)
```

```python
fig, ax = plt.subplots() ## specify number of rows, columns; if blank defaults to 1 by 1

## instead of using plt.plot we use ax.plot
ax.plot(age, salary, label = 'Base Salary')

ax.plot(age, s_masters, label = 'Salary with Masters')

ax.legend()
ax.set_title('Salary by Age')
ax.set_ylabel('Salary')
ax.set_xlabel('Age')
```

```python
## Subplots create a figure and then specify a specific number of rows and columns of axes
## We are stacking vertically, so we have a 1D numpy array containing the list of created Axes.

fig, axs = plt.subplots(nrows=2, ncols=1)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(age, salary, label = 'Base Salary')
axs[1].plot(age, s_masters, label = 'Salary with Masters')
```

```python
## If you are creating just a few Axes, it's handy to unpack them immediately to dedicated variables for each Axes.

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(age, salary, color='#444444',
         linestyle='--', label='Base Salary')  #sharex=True
 
ax2.plot(age, s_masters, color='#444444',
         linestyle='--', label='Salary with Masters')

ax1.legend()
ax1.set_title('Salary by Age')
ax2.set_xlabel('Age')
ax1.set_ylabel('Salary')

ax2.legend()
ax2.set_title('Salary by Age')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary')
```
